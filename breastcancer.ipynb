```python
x=cancer[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']].values
x[0:2]
```

```python
y=cancer['diagnosis'].values
y[0:6]
```

```python
#machine learning
# import from sk learn what you need
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
#random state times zenye hiyo data ilichagua times ama criteria ulichagua ndo itatumia for those  times 
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=7)
```

```python
#training the model
#n_neighbors ni no of neighbors
k=1
pro = KNeighborsClassifier(n_neighbors=k)
pro.fit(x_train, y_train)
```

```python
# hapa mtu anapredict using x_test kujua if prog itaipea hiyo label ilikuwa ie y_test
# so tutacompare y_test na y_pred
y_pred=pro.predict(x_test)
y_pred[0:5]
```

```python
print('Accuracy is', metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
#confusion matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)

sns.heatmap(cm,annot=True)
```

```python
cr = classification_report(y_test,y_pred)
print(cr)
```

```python
accuracy_list={"k":[], "scores":[]}
for k in range(1,21):
    pro = KNeighborsClassifier(n_neighbors=k)
    pro.fit(x_train,y_train)
    y_pred=pro.predict(x_test)
    accuracy_list["k"].append (k)
    accuracy_list["scores"].append(metrics.accuracy_score(y_test,y_pred))
    
accuracy_list
```

```python
k_tester=pd.DataFrame(accuracy_list)
k_tester
```

```python
plt.plot(k_tester["k"], k_tester["scores"])
```

```python
plt.xlabel("k")
plt.ylabel("score")
plt.title("Accuracy")
plt.show()
```
